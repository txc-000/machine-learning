```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("tindak-pidana2021-2023.csv", header=2)
df.columns = ["Kepolisian Daerah", "2021", "2022", "2023"]
df = df.dropna(how="all")
df = df[df["Kepolisian Daerah"] != "INDONESIA"]
tahun = ["2021", "2022", "2023"]
fig, axes = plt.subplots(1, 3, figsize=(18, 6))  # Membuat 3 subplot sejajar

for i, thn in enumerate(tahun):
    ax = axes[i]
    df.set_index("Kepolisian Daerah")[thn].plot.pie(
        autopct='%1.1f%%',
        startangle=0,
        cmap="viridis",
        ax=ax
    )
    ax.set_title(f"Presentase Tindak Pidana {thn}")
    ax.set_ylabel('')

# Menampilkan plot
plt.tight_layout()
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("tindak-pidana2021-2023.csv", header=2)
df.columns = ["Kepolisian Daerah", "2021", "2022", "2023"]
df = df.dropna(how="all")
df = df[df["Kepolisian Daerah"] != "INDONESIA"]
fig, axes = plt.subplots(1, 3, figsize=(18, 6), sharey=True)

# Daftar tahun yang akan dibuat grafiknya
tahun = ["2021", "2022", "2023"]

# Loop untuk membuat line graph per tahun
for i, thn in enumerate(tahun):
    ax = axes[i]

    ax.plot(df["Kepolisian Daerah"], df[thn], marker="o", linestyle="-", color="b", label=thn)

    # Menampilkan jumlah di atas setiap titik
    for j, txt in enumerate(df[thn]):
        ax.text(j, txt + 500, str(txt), fontsize=10, ha="center", color="black")

    # Menyesuaikan tampilan
    ax.set_title(f"Tindak Pidana Tahun {thn}", fontsize=12, fontweight="bold")
    ax.set_xticklabels(df["Kepolisian Daerah"], rotation=90, fontsize=10)
    ax.set_xlabel("Kepolisian Daerah")
    if i == 0:
        ax.set_ylabel("Jumlah Tindak Pidana")

    ax.grid(True, linestyle="--", alpha=0.5)

# Mengatur layout agar tidak bertumpuk
plt.tight_layout()
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("tindak-pidana2021-2023.csv", header=2)
df.columns = ["Kepolisian Daerah", "2021", "2022", "2023"]
df = df.dropna(how="all")
df = df[df["Kepolisian Daerah"] != "INDONESIA"]
fig, axes = plt.subplots(3, 1, figsize=(12, 16), sharex=True)

#warna untuk tiap tahun
colors = ["blue", "green", "red"]
labels = ["2021", "2022", "2023"]

# Loop untuk membuat histogram horizontal per tahun
for i, tahun in enumerate(["2021", "2022", "2023"]):
    ax = axes[i]
    df_sorted = df.sort_values(by=tahun, ascending=True)
    ax.barh(df_sorted["Kepolisian Daerah"], df_sorted[tahun], color=colors[i])
    for index, value in enumerate(df_sorted[tahun]):
        ax.text(value + 200, index, str(value), va="center", fontsize=10, fontweight="bold")
    ax.set_title(f"Jumlah Tindak Pidana per Provinsi ({tahun})", fontsize=14, fontweight="bold", color=colors[i])
    ax.set_ylabel("Provinsi", fontsize=12)

plt.xlabel("Jumlah Tindak Pidana", fontsize=12)
# Menampilkan plot
plt.tight_layout()
plt.show()


```
